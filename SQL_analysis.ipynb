# Detailed analysis of vacancies

1. Lets find out how many vacancies are in each area and sort them.

```python
query_4_1 = f'''select a.name region, count(v.id) qty
                from vacancies v
                    join areas a on a.id=v.area_id
                group by a.name
                order by 2 desc
                limit 5
            '''
```

```python
cnt_area= pd.read_sql_query(query_4_1, connection)
display(cnt_area)
```

2. How many vacancies have filled salary range?

```python
query_4_2 = f'''select count(*)
                from vacancies
                where salary_from is not null or salary_to is not null
            '''
```

2. Finding annomaly area where there are no vacancies but plenty of employers.

```python
query_5_1 = f'''select a.name as name, count(e.id) as emp, count(v.id) as vac
                from areas a
                    left join vacancies v on a.id=v.area_id
                    left join employers e on a.id= e.area
                group by a.name 
                having count(v.id)=0
                order by 2 desc
                limit 1
            '''
```

5. Average quantity of key skills needed for DS vacancies.

```python
query_6_5 = f'''select *
                from vacancies v 
                where (
                    lower(v.name) like '%data scientist%' or 
                    lower(v.name) like '%data science%' or 
                    lower(v.name) like '%исследователь данных%' or 
                    (v.name like '%ML%' and v.name not like '%HTML%') or 
                    lower(v.name) like '%machine learning%' or 
                    lower(v.name) like '%машинн%обучен%'
                )                  
            '''
```

```python
import numpy as np
avg_skills= pd.read_sql_query(query_6_5, connection)
def get_qty(text):
    x=0
    if text is not None:
        text= text.split('\t')
        x+=len(text)
        return x
    else:
        return np.nan
avg_skills['qty']=avg_skills['key_skills'].apply(get_qty)
round(avg_skills['qty'].mean(), 2)  
```

6. Average DS salary.

```python
query_6_6 = f'''select v.experience as experience, avg((coalesce(v.salary_from, 0)+ coalesce(v.salary_to, 0))/
                    (case when v.salary_from is not null and v.salary_to is not null then 2 else 1 end)) as average_salary
                from vacancies v 
                where (
                    lower(v.name) like '%data scientist%' or 
                    lower(v.name) like '%data science%' or 
                    lower(v.name) like '%исследователь данных%' or 
                    (v.name like '%ML%' and v.name not like '%HTML%') or 
                    lower(v.name) like '%machine learning%' or 
                    lower(v.name) like '%машинн%обучен%'
                ) and (
                    v.salary_from is not null or
                    v.salary_to is not null
                    )
                group by v.experience                  
            '''
```

```python
avg_salary= pd.read_sql_query(query_6_6, connection)
display(avg_salary)
```

***

This analysis shows that data science is in high demand, although the barrier to entry for newcomers is significantly higher. This is especially true for data scientists, for whom job openings are significantly fewer and average starting salaries are almost half as much. Some of the most in-demand skills are SQL and Python.
